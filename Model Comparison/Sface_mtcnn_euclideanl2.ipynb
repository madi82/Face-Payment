```python
from sklearn.datasets import fetch_lfw_pairs
```

```python
fetch_lfw_pairs=fetch_lfw_pairs(subset="test", color=True , resize=1)
```

```python
pairs=fetch_lfw_pairs.pairs
targets=fetch_lfw_pairs.target
target_names=fetch_lfw_pairs.target_names
```

```python
actuals=[]
predictions=[]
```

```python
from deepface import DeepFace
```

```python
model=DeepFace.build_model('SFace')
```

```python
import tqdm
predictions=[]
for idx in tqdm.tqdm(range(pairs.shape[0])):

  pair=pairs[idx]
  img1=pair[0]
  img2=pair[1]
  actual=targets[idx]
  
  obj= DeepFace.verify(img1,img2,model_name="SFace", model=model,enforce_detection=False,detector_backend="mtcnn",distance_metric="euclidean_l2")
  prediction=1 if obj["verified"]==True else 0

  predictions.append(prediction)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(targets,predictions))
```
